```python
asus=asus.dropna()
```

```python
data_list=[amd,asus,intel,msi,nvidia]
for data in data_list:
  data['Date']=pd.to_datetime(data['Date'])
```

```python
data_list=[amd,asus,intel,msi,nvidia]
names=[0,1,2,3,4]
index=0
for data in data_list:
  dates=data['Date']
  data['Company']=np.repeat(names[index],len(data))
  data['Year']=dates.dt.year
  data['Month']=dates.dt.month
  data['Day']=dates.dt.day
  index+=1

```

```python
data_list=[amd,asus,intel,msi,nvidia]
test_data=[]
train_data=[]
for data in data_list:
  train= data[:int(0.8*len(data))]
  test=data[int(0.8*len(data)):]
  test_data.append(test)
  train_data.append(train)
  print(test.shape,train.shape)
```

```python
train_data=pd.concat(train_data)
test_data=pd.concat(test_data)
print(train_data.shape)
print(test_data.shape)
```

```python
x_train=train_data[['Open','High','Low','Volume','Year','Month','Day','Company']]
x_test=test_data[['Open','High','Low','Volume','Year','Month','Day','Company']]
y_train=train_data['Close']
y_test=test_data['Close']
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print('Train score',lr.score(x_train,y_train))
print('Test score',lr.score(x_test,y_test))
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2_score=r2_score(y_test,y_pred)
r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,r2_score
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
print('Train score',dt.score(x_train,y_train))
print('Test score',dt.score(x_test,y_test))
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
print('mae',mean_absolute_error(y_test,y_pred))
print('r2_score',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
print('Train score',rf.score(x_train,y_train))
print('Test score',rf.score(x_test,y_test))
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
print ('mae',mean_absolute_error(y_test,y_pred))
print('r2_score',r2_score(y_test,y_pred))
```
